### Подготовка данных  для анализа данных
- Объединение таблиц
- Замена названия столбцов (приведение к нижнему регистру)
- Замена пропусков
- Создание столбца с целевым признаком
- Создание столбца с длительностью договора
- Преобразование данных в нужные типы

```python
#Объединяем таблицы по 'customerID'
dfs = [df_contract, df_personal, df_internet, df_phone]
final_df = reduce(lambda left,right: pd.merge(left,right,on=['customerID'],
                  how='outer'), dfs)
```

```python
#Обзор данных сводной таблицы
final_df.info()
```

- 20 столбцов, 7043 значений
- пропуски в данных в столбцах 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'MultipleLines', заменяем на "No", т.к. предполагаем, что доп. услуги не подключены
- явные дубли не найдены

```python
# Перевод названия столбцов в нижний регистр и разделение нижним подчеркиванием
final_df.columns = ['customer_ID', 'Begin_Date', 'End_Date', 'Type'
                    , 'Paperless_Billing', 'Payment_Method', 'Monthly_Charges'
                    , 'Total_Charges', 'gender', 'Senior_Citizen', 'Partner'
                    , 'Dependents','Internet_Service', 'Online_Security'
                    , 'Online_Backup', 'Device_Protection', 'Tech_Support'
                    , 'Streaming_TV', 'Streaming_Movies', 'Multiple_Lines']

final_df.columns = final_df.columns.str.lower()
```

```python
# Замена пропусков на 'No'
final_df = final_df.replace(np.NaN, 'No')
```

```python
#Выделение целевого признака
final_df['target'] = (final_df['end_date'] != 'No').astype(int)
final_df['target'].value_counts(normalize=True)
```

- Классы в целевом признаке не сбалансированы

```python
#преобразование данных в тип datetime
final_df['begin_date'] = pd.to_datetime(final_df['begin_date'])
final_df['end_date'] = final_df['end_date'].replace(['No'], ['2020-02-01'])
final_df['end_date'] = pd.to_datetime(final_df['end_date'])
```

```python
#Создание признака длительности действия договора
final_df['duration'] = (final_df['end_date'] - final_df['begin_date']).dt.days
```

```python
#Преобразование данных в тип float, int
final_df['total_charges'] = pd.to_numeric(final_df['total_charges']
                                          , errors='coerce')
final_df['senior_citizen']= final_df['senior_citizen'].astype(str)
```

### Подготовка признаков для машинного обучения
- Выделение целевого признака
- Удаление неинформативных и мультколлениарных признаков
- Разделение на тренировочную и тестовую выборки
- Кодирование методом OHE и масштабирование категориальных признаков
- Масштабирование признаков для модели CatBoost

```python
#Выделение целевого признака
y = final_df['target']
#Удаление неинформативных и мультколлениарных признаков
x = final_df.drop(['customer_id', 'begin_date', 'end_date'
                   , 'total_charges', 'target'], axis=1)
```

```python
x.info()
```

```python
#Разделение на тренировочную и тестовую выборки
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25
                                                    , random_state=Random_State
                                                    , stratify=y)
```

```python
category = list(x_train.select_dtypes(['object']).columns)
num_features = ['monthly_charges', 'duration']
```

```python
#Кодирование категориальных признаков методом OHE
x_train_ohe = x_train.copy()
x_test_ohe = x_test.copy()

encoder_ohe = OneHotEncoder(drop='first', handle_unknown='ignore', sparse=False)
encoder_ohe.fit(x_train_ohe[category])
x_train_ohe[encoder_ohe.get_feature_names_out()] = encoder_ohe.transform(x_train_ohe[category])
x_train_ohe = x_train_ohe.drop(category, axis=1)

x_test_ohe[encoder_ohe.get_feature_names_out()] = encoder_ohe.transform(x_test_ohe[category])
x_test_ohe = x_test_ohe.drop(category, axis=1)
print(x_train_ohe.info())

#Признаки для модели RandomForest без  масштабирования

x_train_rf = x_train_ohe.copy()
x_test_rf = x_test_ohe.copy()

#Масштабирование признакоав
scaler = StandardScaler()
x_train_ohe[num_features] = scaler.fit_transform(x_train_ohe[num_features])
x_train_ohe.head(5)

x_test_ohe[num_features] = scaler.transform(x_test_ohe[num_features])
x_test_ohe.head(5)



```

```python
#Масштабирование признакоав для модели CatBoost
x_train_cat = x_train.copy()
x_test_cat = x_test.copy()

scaler = StandardScaler()
x_train_cat[num_features] = scaler.fit_transform(x_train_cat[num_features])
x_train_cat.head(5)

x_test_cat[num_features] = scaler.transform(x_test_cat[num_features])
x_test_cat.head(5)

#Подготовка признаков для модели CatBoost без кодирования и масштабирования
x_train_cat_alt = x_train.copy()
x_test_cat_alt = x_test.copy()
```

### Подбор параметров и выбор наилучшей модели машинного обучения для задачи классификации
- Модель RandomForest
- Модель LogisticRegression
- Модель CatBoostClassifier

```python
# Подбор гиперпараметров для модели Решающий лес
model = RandomForestClassifier(random_state=Random_State, class_weight='balanced')
parametry = [{'n_estimators': range(30, 60, 5),
              'max_depth': range(2, 10, 2)}]
grid_rfc = GridSearchCV(model
                        , parametry                        
                        , cv=CV
                        , n_jobs = -1
                        , scoring='roc_auc')
grid_rfc.fit(x_train_ohe, y_train)
rfc_model=grid_rfc.best_estimator_
print('Параметры', grid_rfc.best_params_)
# Accuracy:
accuracy_rf = accuracy_score(y_train, rfc_model.predict(x_train_ohe))
# Вывод результата
results_rf = pd.DataFrame({'RandomForestClassifier': [round(grid_rfc.best_score_, 2),                                          
                                                      round(accuracy_rf, 2)]}
                          , index=['AUC-ROC', 'Accuracy'])
display(results_rf)
```

```python
# Подбор гиперпараметров для модели Решающий лес на выборке без масштабирования
model = RandomForestClassifier(random_state=Random_State, class_weight='balanced')
parametry = [{'n_estimators': range(30, 60, 5),
              'max_depth': range(2, 10, 2)}]
grid_rfc_alt = GridSearchCV(model
                        , parametry                        
                        , cv=CV
                        , n_jobs = -1
                        , scoring='roc_auc')
grid_rfc_alt.fit(x_train_rf, y_train)
rfc_model_alt=grid_rfc_alt.best_estimator_
print('Параметры', grid_rfc_alt.best_params_)
# Accuracy:
accuracy_rf_alt = accuracy_score(y_train, rfc_model_alt.predict(x_train_rf))
# Вывод результата
results_rf_alt = pd.DataFrame({'RandomForestClassifier_alt': [round(grid_rfc_alt.best_score_, 2),                                          
                                                      round(accuracy_rf_alt, 2)]}
                          , index=['AUC-ROC', 'Accuracy'])
display(results_rf_alt)
```

```python
# Подбор гиперпараметров для модели Логистическая регрессия
model= LogisticRegression(random_state=Random_State
                          , solver='liblinear'
                          , class_weight='balanced')
parametry = [{'C': range (1, 5)}]
grid_lr = GridSearchCV(model, parametry, scoring='roc_auc', cv=CV, n_jobs = -1)
grid_lr.fit(x_train_ohe, y_train)
lr_model=grid_lr.best_estimator_
print('Параметры:', grid_lr.best_params_)

# Accuracy:
accuracy_lr = accuracy_score(y_train, lr_model.predict(x_train_ohe))

# Вывод результата
results_lr = pd.DataFrame({'LogisticRegression': [round(grid_lr.best_score_, 2),                                         
                                                  round(accuracy_lr, 2)]}
                          , index=['AUC-ROC', 'Accuracy'])
display(results_lr)
```

```python
# Подбор гиперпараметров для модели CatBoostClassifier
model = CatBoostClassifier(auto_class_weights='Balanced'
                           , random_state=Random_State
                           , verbose=False
                           , allow_writing_files=False
                           , cat_features=category)
parametry = [{'depth': range(2, 6),
              'iterations': range(250, 351, 50),
              'learning_rate': [0.1, 0.15]}]
grid_cbc = GridSearchCV(model, parametry, scoring='roc_auc', cv=CV, n_jobs = -1)
grid_cbc.fit(x_train_cat, y_train)
cbc_model=grid_cbc.best_estimator_
print('Параметры', grid_cbc.best_params_)

# Accuracy:
accuracy_cbc = accuracy_score(y_train, cbc_model.predict(x_train_cat))

# Вывод результата
results_cbc = pd.DataFrame({'CatBoostClassifier': [round(grid_cbc.best_score_, 2),                                          
                                                   round(accuracy_cbc, 2)]}
                           , index=['AUC-ROC', 'Accuracy'])
display(results_cbc)
```

```python
# Подбор гиперпараметров для модели CatBoostClassifier 
# без масштабирования признаков
model = CatBoostClassifier(auto_class_weights='Balanced'
                           , random_state=Random_State
                           , verbose=False
                           , allow_writing_files=False
                           , cat_features=category)
parametry = [{'depth': range(2, 6),
              'iterations': range(250, 351, 50),
              'learning_rate': [0.1, 0.15]}]
grid_cbc_alt = GridSearchCV(model, parametry, scoring='roc_auc', cv=CV, n_jobs = -1)
grid_cbc_alt.fit(x_train_cat_alt, y_train)
cbc_model_alt=grid_cbc_alt.best_estimator_
print('Параметры', grid_cbc_alt.best_params_)

# Accuracy:
accuracy_cbc_alt = accuracy_score(y_train, cbc_model.predict(x_train_cat_alt))

# Вывод результата
results_cbc_alt = pd.DataFrame({'CatBoostClassifier_alt': [round(grid_cbc_alt.best_score_, 2),                                          
                                                   round(accuracy_cbc_alt, 2)]}
                           , index=['AUC-ROC', 'Accuracy'])
display(results_cbc_alt)
```
